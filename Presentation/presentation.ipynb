## Example of the SAT problem. 

```python
def sat(x1,x2):
    return (x1 or not x2) and (not x1 or x2)
```

```python
import itertools
import time
t0 = time.time()
print(t0)
for (x1, x2) in list(itertools.product([False, True], repeat=2)):
    print("Input 1 :", x1, ", Input 2 :", x2, '| Result ->', sat(x1,x2))
print("The calculation time for 2 variables is", time.time()-t0)
```

#  <font color = 'red'> Combinatorial Problem: </font>
We have three characters a,b,c.  generate all possible combinations of 
‘a’ , ‘b’ and ‘c’ 

```python
###  Recursive Solution: 
# Define a function which generates all possible combinations of a given array of elements
def combinations(elements):
    if elements:
      result = combinations(elements[:-1])
      return result + [c + [elements[-1]] for c in result]
    else:
      return [[]]

print(combinations(['a','b','c']))
```

```python
### Measure and Plot Computation Times Against Character Pool Length
import numpy as np
import matplotlib.pyplot as plt

def plotCalTime(num_chars, cal_time):	
	fig = plt.figure()
	ax = plt.axes()
	plt.grid()

	x = num_chars
	y = cal_time 
	plt.plot(x, y, 'r--', linewidth=3, marker=".", mfc='g', markersize=20)
	plt.xlabel('Numbers of Characters', fontsize=15)
	plt.ylabel('Computation Time (ms)', fontsize=15)
	plt.show()


# Generate a list of the alphabetic characters from 'a' to 'm'
character_pool = [x for x in range(ord('a'), ord('w')+1)]
cal_time_recurse = []
num_combine_recurse = []
# Measure the time taken to generate combinations for a range of lengths of the character array
for num_char in range(1,len(character_pool)+1):
	t0 = time.time()
	combinations(character_pool[:num_char])
	cal_time_recurse.append((time.time()-t0)*1000)
	num_combine_recurse.append(num_char)

plotCalTime(num_combine_recurse, cal_time_recurse)
### Note that, apart from possible anomalies, the computation time increases exponentially with added characters
```

```python
### The generation of combinations can be mapped to an SAT problem by representing combinations as boolean arrays encoding the existence or not of each element
cal_time_SAT = []
num_combine_SAT = []
for num_chars in range(1,len(character_pool)+1):	
	t0 = time.time()
	# Leverage itertools to generate the boolean map
	existence_map = list(itertools.product([False, True], repeat=num_chars))
	cal_time_SAT.append((time.time()-t0)*1000)
	num_combine_SAT.append(num_chars)

plotCalTime(num_combine_SAT, cal_time_SAT)
### Note there is still an exponential relationship, but the computation completes much more quickly
```

```python
### We can then define a function which maps SAT problem outputs to combinatorial problem outputs
### In this example, the boolean values are mapped to lists of ordered alphabetic characters
def SAT(existence_map): # `existence_map` is a map representing the existence or not of each character in each element of the combination
	combinations = []
	for ele in existence_map:
		total_string = ""
		for index, char in enumerate(ele):
			# If the element exists in the combination, write it to the string
			if char == True:
				char_string = chr(index+97) # Map the position in the boolean map to alphabetic characters
				total_string += char_string
		combinations.append(total_string)
	return combinations
print(SAT(list(itertools.product([False, True], repeat=3))))
```

```python
### For Illustration, a simple method to reconstruct combinations of 'a','b','c' from the existence map
def simple_SAT(x1, x2, x3):
    if x1 == True:
        stringa = 'a'
    else:
        stringa = ''
    if x2 == True:
        stringb = 'b'
    else:
        stringb = ''
    if x3 == True:
        stringc = 'c'
    else:
        stringc = ''
    return stringa + stringb + stringc
	
print([simple_SAT(x1,x2,x3) for x1,x2,x3 in list(itertools.product([False, True], repeat=3))])
```
